```python
#simulating the main loop
import time

def getting_vis():
    print("vision started")
    time.sleep(0.5)#simulating execution of the function
    actual,angle=([0.00,1.00],45) #fake data
    return (actual, angle)
    
def kalman(actual, angle):
    print("kalman started")
    time.sleep(0.1) #simulating execution of the function
    print("kalman received position {} and angle {}".format(actual,angle))
    pred_pos=[actual[0]+0.1,actual[1]-0.1]
    pred_angle= angle/1,1
    return pred_pos, pred_angle

def astolfi():
    print("astolfi started")
    time.sleep(0.1)
    
def local_detection():
    print("local detection started")
    for i in range(3):
        prox=ctrl.read_prox_sensors(node,client)
        print(prox)
        if (prox[2]>1000):
            print("object detected")
        aw(client.sleep(0.2))
    
def main_loop():
    while True:
        local_detection()
        actual, angle= getting_vis()
        pred_pos, pred_angle = kalman(actual, angle)
        astolfi()
    
    
#main_loop()
```

```python
def local_nav(prox_horizontal,y):
    leds_top = [30,30,30]
    
    # obstacle avoidance: ANN
    w_l = [40,  20, -20, -20, -40,  30, -10, 8, 0]
    w_r = [-40, -20, -20,  20,  40, -10, 30, 0, 8]

    # Scale factors for sensors and constant factor
    sensor_scale = 200
    constant_scale = 20

    x = [0,0,0,0,0,0,0,0,0]

    # Memory
    x[7] = y[0]//20
    x[8] = y[1]//20

    for i in range(7):
        # Get and scale inputs
        x[i] = prox_horizontal[i] // sensor_scale

    y = [0,0]    
    for i in range(len(x)):    
        # Compute outputs of neurons and set motor powers
        y[0] = y[0] + x[i] * w_l[i]
        y[1] = y[1] + x[i] * w_r[i]
        y[0] = y[0] 
        y[1] = y[1]
    
    return y

def update_state(state,obstThrL,obstThrH):
    if state == 0: 
        # switch from goal tracking to obst avoidance if obstacle detected
        if (obst[0] > obstThrH):
            state = 1
            print("change state to local")
        elif (obst[2] > obstThrH):
            state = 1
            print("change state to local")
        elif (obst[1] > obstThrH):
            state = 1
            print("change state to local")
    elif state == 1:
        #if obst[1] < obstThrL:
        if obst[0] < obstThrL:
            if obst[2] < obstThrL:
                # switch from obst avoidance to goal tracking if obstacle got unseen
                state = 0
                print("change state to global")
    
    return state

def new_glob_nav(checkpoints,starting_angle,node,client):
    angle=starting_angle
    for i in range(len(checkpoints)-1):   
        navigate(checkpoints[i+1], checkpoints[i], angle, node, client)
        angle=angle2points(checkpoints[i+1],checkpoints[i],node)
    stopmotors(node)
    
#Initializing gloabl path planning
p1 = [(1, 0.5), (2, 0.5), (2, 1.5), (1, 1.5)]
p3 = [(3, 3), (4, 3), (4, 4), (3, 4)]
list_obstacles=[p1,p3]
start_point=[0.0,0.0]
end_point=[4.5, 3.5]
margin=0.2
start_angle=0

path = glob.global_pathplanning(start_point,end_point,list_obstacles,margin)
checkpoints=glob.geometric_path_to_vector(path)  #can be accesed via g too
print("Checkpoints",checkpoints)

checkp=0
actual_angle=0  

y = [50,50]          # powers given to motor wheels with global nav, here TEST 0,0
state = 0          # 0=global nav, 1=local nav
obst = [0,0,0]     # measurements from left, middle, right prox sensors

obstThrL = 2500      # low obstacle threshold to switch state 1->0
obstThrL2 = 3000
obstThrH = 4000      # high obstacle threshold to switch state 0->1  

#start_time = time.time()
#seconds = 1

while True:
    
    # get the proximity sensors value and obstacles values
    prox_horizontal = ctrl.read_prox_sensors(node,client)
    obst = [prox_horizontal[0], prox_horizontal[2], prox_horizontal[4]]
    
    # update the state and speed motors values
    state = update_state(state,obstThrL,obstThrH)
    y = ctrl.read_motors_speed(node,client)
                
    #current_time = time.time()
    #elapsed_time = current_time - start_time
    #if elapsed_time > seconds:
        #print("Finished iterating in: " + str(int(elapsed_time))  + " seconds")
        #break
                
    print("state", state)

    if state == 0:
        """
        # When vision and kalman ready
        
        actual,angle= getting_vis()
        pred_pos,pred_angle = kalman(actual, angle)
        #turn(next,actual,actual_angle, node, client)
        #forward(next,actual, node, client)
        #angle=angle2points(checkpoints[i+1],checkpoints[i],node)
        astolfi(actual_pos, goal_pos, actual_angle, node)
        """
        actual_pos=checkpoints[checkp]
        goal_pos=checkpoints[checkp+1]
        
        # Mehtof
        #y = ctrl2.turn(goal_pos, actual_pos,actual_angle, node, client)
        #ctrl.set_motors(y[0],y[1],node)
        #y = ctrl2.forward(goal_pos,actual_pos, node, client)
        #actual_angle = ctrl2.angle2points(goal_pos,actual_pos,node)
        
        #actual_angle = ctrl2.angle2points(goal_pos,actual_pos,node)
        y = ctrl2.astolfi(actual_pos, goal_pos, actual_angle, node)
        print("y astolfi ",y)
        
        #y = [50,50]
        
        #iterating through checkpoints 
        #ctrl.globnav(checkpoints,start_angle,node,client)
        
        ctrl.set_motors(y[0],y[1],node)

    elif state == 1:
        print(obst)
        y = local_nav(prox_horizontal,y)
        print(y)
        ctrl.set_motors(y[0],y[1],node)

ctrl.stopmotors(node)
```
