```python
def playVideo(): 
    # 1. Create an object. Zero for external camera
    video = cv2.VideoCapture(0)

    a = 0
    video_imgs = []

    while True:
        a = a + 1
        # 3. Create a frame object
        check, frame = video.read()

        #print(check)
        print(frame) #representing image
        
        # all frames
        video_imgs.append(frame)
            
        # 4. Show the frame!
        cv2.imshow("Capturing",frame)

        # 5. for press any key to out (milisec)
        #cv2.waitKey(0)

        # 7. for playing
        key = cv2.waitKey(1)

        if key == ord('q') :
            break

    # 2. Shutdown camera
    video.release()

    cv2.destroyAllWindows
    
    return frame
                  
```

```python
def takePicture(): 
    # 1. Create an object. Zero for external camera
    video = cv2.VideoCapture(0)

    # 3. Create a frame object
    check, frame = video.read()
    
    #cv2.imwrite(frame, output)

    #print(check)
    print(frame) #representing image
        
    # 4. Show the frame!
    cv2.imshow("Capturing picture",frame)

    # 5. for press any key to out (milisec)
    cv2.waitKey(0)

    # 2. Shutdown camera
    video.release()
    return frame
```
