```python
import cv2, time
import matplotlib.pyplot as plt
import numpy as np
import Vision as vis
```

```python
env = cv2.imread('env_final.jpeg')
env = cv2.cvtColor(env, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(env)
plt.show()
```

```python
new_env = vis.img_calibration(env)
plt.figure()
plt.imshow(new_env)
plt.show()
```

```python
goal = vis.detectCircle(new_env,'goal')
start = vis.detectCircle(new_env,'start')
print(start[0])
```

```python
lineThickness = 2
x1 = start[0] 
y1 = start[1]
x2 = goal[0]
y2 = goal[1]
cv2.line(new_env, (x1, y1), (x2, y2), (0,255,0), lineThickness)
cv2.imshow("results", new_env)
if cv2.waitKey(1) & 0xFF == ord('q'):      
    break
```
